```python
# Load Libraries
import pandas as pd
import numpy as np
```

```python
# Creating a series from a list
data = [10,20,30,40,50]
series_from_list = pd.Series(data) # series is name of Function
print(series_from_list)
print(type(series_from_list)) # type is series
```

```python
numpy_array = np.array([1,2,3,4,5])
series_from_numpy = pd.Series(numpy_array)
print(series_from_numpy)

numpy_array = np.array(data)
series_from_numpy = pd.Series(numpy_array)
print(series_from_numpy)
```

* custom index

```python
# Creating a series from a list
data = [10,20,30,40,50]
custom_index = pd.Series(data , index=['A','B','C','D','E']) # custom indexc
custom_index
```

* Accessing elements by position (zero based index)

```python
# Access 30 from variable series_from_list
index_value = series_from_list[2]
print(index_value)

# Access 40 from variable custom_index
index_value = custom_index['D']
print(index_value)
```

```python
slice_series = series_from_list[1:4]
print(slice_series)
```
