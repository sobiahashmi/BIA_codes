# Feature Encoding

Feature encoding in Python is the process of converting `Categorical Data` into `Numerical Format` so that it can be used in machine learning models. This is crucial because most machine learning algorithms work only with numerical data. Following we will be discussed:

- Label encoding
- Ordinal encoding
- One hot encoding
- Binary encoding

```python
# Load Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,OneHotEncoder
from category_encoders import BinaryEncoder
```

```python
# Load Dataet
df = sns.load_dataset('tips')
df.head()
```

```python
# check valuecounts
print(df['sex'].value_counts())
print(df['smoker'].value_counts())

df['day'].value_counts(), df['time'].value_counts()
```

## 1. Label Encoder 

* **Let's encode the time with Label encoder in sklearn**

```python
le = LabelEncoder()
```

```python
# LabelEncoder is applied to `time` and save into new variable 
df['encoded_time'] = le.fit_transform(df['time'])
```

```python
df['encoded_time'].value_counts()
```

## 2. Ordinal Encoder 

* **Let's encode the day with ordinal encoder using specific order**

```python
df['day'].value_counts()
```

```python
oe = OrdinalEncoder(categories=[['Sat','Sun','Thur','Fri']])
df['encoded_day'] = oe.fit_transform(df[['day']])
```

```python
df.sample(20)
```

## 3. One Hot Encoder 

* **Let's encode the day with One Hot Encoding on `Day` columns**

```python
ohe = OneHotEncoder()
ohe.fit_transform(df[['smoker']]).toarray()
```
