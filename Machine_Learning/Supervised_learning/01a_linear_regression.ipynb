Linear regression is a simple algorithm used to predict a target value based on one or more input features by fitting a straight line (or hyperplane in higher dimensions) to the data. It works by minimizing the difference between the predicted values and the actual values using a method called least squares. âš¡ðŸ’¡

There are two main types:
- Simple Regression - only one independent variable (e.g. sales)
- Multiple Regression - more than one independent variable (e.g. sales, price)-

```python
# Load libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
""
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error , r2_score
```

<a id="13"></a>
<h1 style="background-color:#435420;font-family:newtimeroman;font-size:300%;text-align:center;border-radius: 15px 50px;color:#FF9900;">Step-2 Load Librariesâ€‹</h1>
<figcaption style="text-align: center;">
    <strong>
    </strong>
</figcaption>

```python
# Load Dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# scatter plot of total_bill and tip
sns.scatterplot(x='total_bill' , y='tip' , data=df)
```

<a id="13"></a>
<h1 style="background-color:#435420;font-family:newtimeroman;font-size:300%;text-align:center;border-radius: 15px 50px;color:#FF9900;">Step-3 Data Splittingâ€‹</h1>
<figcaption style="text-align: center;">
    <strong>
    </strong>
</figcaption>

```python
# split the data into X and y
X = df[['total_bill']]
y = df['tip']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=32)
```

<a id="13"></a>
<h1 style="background-color:#435420;font-family:newtimeroman;font-size:300%;text-align:center;border-radius: 15px 50px;color:#FF9900;">Step-4 Linear Regression Modelâ€‹</h1>
<figcaption style="text-align: center;">
    <strong>
    </strong>
</figcaption>

```python
# call the model
model = LinearRegression()
```

```python
# train the model
model.fit(X_train,y_train)
```

<a id="13"></a>
<h1 style="background-color:#435420;font-family:newtimeroman;font-size:300%;text-align:center;border-radius: 15px 50px;color:#FF9900;">Step-5 Data Predictionâ€‹</h1>
<figcaption style="text-align: center;">
    <strong>
    </strong>
</figcaption>

```python
# predict
y_pred = model.predict(X_test)
y_pred
```

```python
# takeout model intercept and slope, make an equation
print('Intercept:', model.intercept_)
print('Coefficiant:', model.coef_)
print('y= ' ,model.intercept_ , '+' , model.coef_ , '* X')
```

<a id="13"></a>
<h1 style="background-color:#435420;font-family:newtimeroman;font-size:300%;text-align:center;border-radius: 15px 50px;color:#FF9900;">Step-6 Model Evaluationâ€‹</h1>
<figcaption style="text-align: center;">
    <strong>
    </strong>
</figcaption>

```python
# evaluate the model
print('MSE: ', mean_squared_error(y_test,y_pred))
print('R2 = ', r2_score(y_test,y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test,y_pred) ))
```

**Insights:**

- 23% r-square shows this model didn't performed well with iris dataset

<a id="13"></a>
<h1 style="background-color:#435420;font-family:newtimeroman;font-size:300%;text-align:center;border-radius: 15px 50px;color:#FF9900;">Step-7 Scatter Plot in 'Original vs Predicted' dataâ€‹</h1>
<figcaption style="text-align: center;">
    <strong>
    </strong>
</figcaption>

```python
# plot the model and data 'Original vs Predicted'
plt.scatter(X_test,y_test)
plt.plot(X_test , y_pred , color = 'red')
```

**Insights:**

* Blue dots shows Original dataset points
* Red line shows best fitted line (of Predicted data)
* If datapoints are so much far from red line
    * Mean Square Error (MSE) will be high.
    * Accuracy will be LOW
* If datapoints are close to red line
    * Mean Square Error (MSE) will be LOW.
    * Accuracy will be HIGH
