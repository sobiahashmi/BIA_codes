### **Subset dataframe**

```python
# Load Libraries
import pandas as pd
import seaborn as sns

# Load Dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# Method 1
filter_pclass = df[df['pclass'].isin([2,3])]
filter_pclass
```

```python
# Method 2
df[(df['pclass'] == 2) | (df['pclass'] == 3)]
```

### **not NULL**

```python
df[df['age'].notnull()] # Age colunm with no NULL values
```

```python
# This will also work with notna
df_known_age = df[df['age'].notna()]
df_known_age
```

```python
df_unknown_age = df[df['age'].isna()]
df_unknown_age
```

```python
# Code will also work with isnull
df_unknown_age = df[df['age'].isnull()]
df_unknown_age
```
