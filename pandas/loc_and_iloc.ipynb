## **'LOC' and 'iLOC'** in panads for selecting data
    * LOC - Selects data by label (row/column names).
      - df.loc['John', 'Score']
  
    * iLOC - Selects data by index (row/column positions).
      - df.iloc[2, 3]

## When to use LOC and iLOC?
  + Use loc when you know the labels (names) of rows or columns.
    - df.loc[0:5,[column1,column2]]
    
  + Use iloc when you know the index positions (numbers) of rows or columns.
    - df.iloc[0:5,0:2]
